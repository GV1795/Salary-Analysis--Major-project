```python
# import the import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the dataset.
df = pd.read_csv("D:/GAURAV NOTES/notes 1965/DataSet/Salary _analysis/ds_salaries new.csv")
df.head()
```

```python
# Display basic information
print(df.info())
```

```python
# Display descriptive statistics
print(df.describe())
```

```python
# Check the shape of the dataset
df.shape
```

## Explore the distribution of salaries for each company.

```python
# Group by 'Company' and calculate descriptive statistics
salary_stats = df.groupby('job_title')['salary_in_usd'].describe()

# Display the salary distribution for each company
print("\nSalary Distribution by job_title:")
print(salary_stats)
```

```python


# Set the style of seaborn
sns.set(style="whitegrid")

# Plot the distribution of salaries for each company
plt.figure(figsize=(10, 8))
sns.boxplot(x="company_location", y="salary_in_usd", data=df)
plt.title('Distribution of Salaries for Each Company')
plt.xlabel('Company Location')
plt.xticks(rotation=90)
plt.ylabel('Salary in USD')
plt.show()
```

### relationship between experience level and salary

```python
# Set up Seaborn style
sns.set(style="whitegrid")

# Create a scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='experience_level', y='salary_in_usd', data=df, hue='experience_level', palette='viridis', s=100)
plt.title('Relationship Between Experience Level and Salary')
plt.xlabel('Experience Level')
plt.ylabel('Salary (USD)')
plt.legend(title='Experience Level', loc='upper right')
plt.show()
```

### Analyze the remote work ratio across different companies.

```python
# Plot the remote work ratio for each company
plt.figure(figsize=(12, 8))
sns.barplot(x="company_location", y="remote_ratio", data=df)
plt.title('Remote Work Ratio Across Different Companies')
plt.xlabel('Company Location')
plt.xticks(rotation=90)
plt.ylabel('Remote Work Ratio')
plt.show()
```

#### Plot the distribution of salaries based on employment type


```python
plt.figure(figsize=(10, 8))
sns.boxplot(x="employment_type", y="salary_in_usd", data=df)
plt.title('Distribution of Salaries Based on Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.show()
```

### Visualize the distribution of salaries using a histogram.


```python
# Plot a histogram of salaries
plt.figure(figsize=(12, 8))
sns.histplot(df['salary_in_usd'], bins=30, kde=True)
plt.title('Distribution of Salaries')
plt.xlabel('Salary in USD')
plt.ylabel('Frequency')
plt.show()
```

### Show a pair plot to explore relationships between numerical variables.

```python
# Create a pair plot for numerical variables
num_vars = ['salary_in_usd', 'remote_ratio']
sns.pairplot(df[num_vars])
plt.suptitle('Pair Plot of Numerical Variables')
plt.show()
```

### Explore the distribution of salaries based on job titles using a violin plot.

```python
# Plot a violin plot for job titles and salaries


# Set a color palette for the plot
palette = "Set3"  # You can choose any seaborn color palette

plt.figure(figsize=(14, 10))
sns.violinplot(x="job_title", y="salary_in_usd", data=df, palette=palette)
plt.title('Distribution of Salaries Based on Job Titles')
plt.xlabel('Job Title')
plt.ylabel('Salary in USD')
plt.xticks(rotation=90, ha='right')

# Add labels for better understanding
plt.xlabel('Job Title', fontsize=14)
plt.ylabel('Salary in USD', fontsize=14)

plt.show()

```

### Visualize the distribution of company sizes using a count plot.

```python
# Plot a count plot for company sizes
plt.figure(figsize=(10, 6))
sns.countplot(x="company_size", data=df)
plt.title('Distribution of Company Sizes')
plt.xlabel('Company Size')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Explore the relationship between years of experience and salary using a scatter plot.

```python
# Plot a scatter plot for experience level and salary
plt.figure(figsize=(12, 8))
sns.scatterplot(x="work_year", y="salary_in_usd", hue="experience_level", data=df)
plt.title('Relationship Between Years of Experience and Salary')
plt.xlabel('Work Year')
plt.ylabel('Salary in USD')
plt.legend(title='Experience Level')
plt.show()
```

### Visualize the distribution of salaries across different locations using a bar chart.

```python
# Plot a bar chart for average salaries across different locations
avg_salary_by_location = df.groupby('company_location')['salary_in_usd'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 8))
avg_salary_by_location.plot(kind='bar', color='skyblue')
plt.title('Average Salaries Across Different Locations')
plt.xlabel('Company Location')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=90, ha='right')
plt.show()
```

### Explore the distribution of salaries based on both experience level and employment type using a heatmap.

```python
# Create a pivot table for salary with experience level and employment type
pivot_table = df.pivot_table(index='experience_level', columns='employment_type', values='salary_in_usd', aggfunc='mean')

# Plot a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(pivot_table, cmap='YlGnBu', annot=True, fmt='.0f', cbar_kws={'label': 'Average Salary in USD'})
plt.title('Average Salaries Based on Experience Level and Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Experience Level')
plt.show()
```

### Visualize the distribution of remote work ratios based on company size using a box plot.

```python
# Plot a box plot for remote work ratio and company size
plt.figure(figsize=(12, 8))
sns.boxplot(x="company_size", y="remote_ratio", data=df)
plt.title('Distribution of Remote Work Ratio Based on Company Size')
plt.xlabel('Company Size')
plt.ylabel('Remote Work Ratio')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Explore the distribution of salaries for each year using a line plot.

```python
# Plot a line plot for the distribution of salaries over the years
plt.figure(figsize=(10, 6))
sns.lineplot(x="work_year", y="salary_in_usd", data=df, estimator='mean', ci=None)
plt.title('Distribution of Salaries Over the Years')
plt.xlabel('Work Year')
plt.ylabel('Average Salary in USD')
plt.show()
```

### Explore the relationship between company size and average salary using a bar plot.

```python
# Plot a bar plot for average salary based on company size
avg_salary_by_company_size = df.groupby('company_size')['salary_in_usd'].mean().sort_values(ascending=False)
bar_width = 0.2
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_salary_by_company_size.index, y=avg_salary_by_company_size.values, palette='viridis')
plt.title('Average Salary Based on Company Size')
plt.xlabel('Company Size')
plt.ylabel('Average Salary in USD')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Visualize the distribution of salaries for each employment type using a violin plot.

```python
# Plot a violin plot for employment type and salaries
plt.figure(figsize=(12, 8))
sns.violinplot(x="employment_type", y="salary_in_usd", data=df, palette='Set2')
plt.title('Distribution of Salaries for Each Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.show()
```

### Explore the distribution of remote work ratios using a pie chart.

```python
# Plot a pie chart for the distribution of remote work ratios
remote_ratio_counts = df['remote_ratio'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(remote_ratio_counts, labels=remote_ratio_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'))
plt.title('Distribution of Remote Work Ratios')
plt.show()
```

### Visualize the distribution of salaries over the years using a line plot with hue for different experience levels.

```python
# Plot a line plot with hue for experience level
plt.figure(figsize=(8, 6))
sns.lineplot(x="work_year", y="salary_in_usd", hue="experience_level", data=df, ci=None)
plt.title('Distribution of Salaries Over the Years by Experience Level')
plt.xlabel('Work Year')
plt.ylabel('Average Salary in USD')
plt.show()
```

### Visualize the distribution of remote work ratios using a donut chart.

```python
# Plot a donut chart for the distribution of remote work ratios
remote_ratio_counts = df['remote_ratio'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(remote_ratio_counts, labels=remote_ratio_counts.index, autopct='%1.1f%%', startangle=90, colors=sns.color_palette('pastel'), wedgeprops=dict(width=0.4))
plt.title('Distribution of Remote Work Ratios (Donut Chart)')
plt.show()
```

### Explore the correlation between numerical variables using a heatmap.

```python
# Plot a heatmap for correlation between numerical variables
correlation_matrix = df[['salary_in_usd', 'remote_ratio']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation Between Numerical Variables')
plt.show()
```

### Create an interactive line chart for the distribution of salaries over the years with experience level as hover text.

```python
import plotly.express as px

# Create an interactive line chart
fig = px.line(df, x="work_year", y="salary_in_usd", color="experience_level", title='Distribution of Salaries Over the Years with Experience Level',
              labels={'salary_in_usd': 'Average Salary in USD', 'work_year': 'Work Year'},
              hover_data={'experience_level': True})
fig.show()
```

### Generate an interactive scatter plot for the relationship between years of experience and salary, with job title as hover text.

```python
# Create an interactive scatter plot
fig = px.scatter(df, x="work_year", y="salary_in_usd", color="job_title", title='Relationship Between Years of Experience and Salary with Job Title',
                 labels={'salary_in_usd': 'Salary in USD', 'work_year': 'Work Year'},
                 hover_data={'job_title': True})
fig.show()
```

### Develop an interactive pie chart for the distribution of remote work ratios.

```python
# Create an interactive pie chart
fig = px.pie(df, names='remote_ratio', title='Distribution of Remote Work Ratios',
             labels={'remote_ratio': 'Remote Work Ratio'})
fig.show()
```

### Generate an interactive 3D scatter plot for the relationship between years of experience, salary, and remote work ratio.

```python
# Create an interactive 3D scatter plot
fig = px.scatter_3d(df, x='work_year', y='salary_in_usd', z='remote_ratio', color='experience_level',
                    title='3D Scatter Plot: Years of Experience, Salary, and Remote Work Ratio',
                    labels={'salary_in_usd': 'Salary in USD', 'work_year': 'Work Year', 'remote_ratio': 'Remote Work Ratio'})
fig.show()
```

### Build an animated bubble chart showing the distribution of salaries over the years, with bubble size representing the company size.

```python
# Create an animated bubble chart
fig = px.scatter(df, x="work_year", y="salary_in_usd", animation_frame="company_size", size="salary_in_usd", color="company_size",
                 title='Animated Bubble Chart: Salary Distribution Over the Years by Company Size',
                 labels={'salary_in_usd': 'Salary in USD', 'work_year': 'Work Year', 'company_size': 'Company Size'})
fig.show()
```

### Develop a choropleth map to visualize the average salaries across different company locations.

```python
# Create a choropleth map
avg_salary_by_location = df.groupby('company_location')['salary_in_usd'].mean().reset_index()
fig = px.choropleth(avg_salary_by_location, locations='company_location', locationmode='country names', color='salary_in_usd',
                    title='Choropleth Map: Average Salaries Across Different Company Locations',
                    labels={'salary_in_usd': 'Average Salary in USD'})
fig.show()
```

### Create an animated sunburst chart to explore the hierarchy of average salaries based on job titles and experience levels.

```python
# Create an animated sunburst chart
fig = px.sunburst(df, path=['job_title', 'experience_level'], values='salary_in_usd',
                  title='Animated Sunburst Chart: Hierarchy of Average Salaries by Job Titles and Experience Levels')
fig.show()
```
